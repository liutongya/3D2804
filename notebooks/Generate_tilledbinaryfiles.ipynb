```python
def gen_binfile(data, savename, ny, nx, 
                ddir='/rigel/ocp/users/tu2140/MITgcm/Southern-Ocean/'):
#     fname = op.join(ddir,fname)
#     data = np.fromfile(fname, dtype=dtypes).astype(np.float64)
    N = data.shape
    if len(N) == 1:
        np.broadcast_to(data[:, None, None], 
                       (N[0], ny, nx)).tofile(op.join(ddir, savename))
    elif len(N) == 2:
        np.broadcast_to(data[:,:, None, None], 
                       (N[0], N[1], ny, nx)).tofile(op.join(ddir, savename))
```

```python
data_1d = np.ma.masked_invalid(np.nanmean(data,axis=0))
polyfit_coeff = np.polyfit(np.ma.masked_array(-z, data_1d.mask).compressed(), 
                       data_1d.compressed(), 7)
print(polyfit_coeff)

poly_ni = (polyfit_coeff[0]*(-z)**7+polyfit_coeff[1]*(-z)**6
            +polyfit_coeff[2]*(-z)**5+polyfit_coeff[3]*(-z)**4
            +polyfit_coeff[4]*(-z)**3+polyfit_coeff[5]*(-z)**2
            +polyfit_coeff[6]*(-z)
            +polyfit_coeff[-1])
```
